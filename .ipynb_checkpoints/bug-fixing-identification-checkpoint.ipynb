```python
from pydriller import Repository
import csv
import os.path
import pandas as pd
from jira import JIRA
```

```python
repoPath = 'https://github.com/apache/ActiveMQ'
commitsFile = 'commits.csv'
commits = []

if not(os.path.isfile(commitsFile)) :
    with open(commitsFile, 'a') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=['hash', 'msg', 'deletions', 'insertions'])
        writer.writeheader()                        
        for commit in Repository(repoPath).traverse_commits():
            record = {'hash': commit.hash, 'msg': commit.msg, 'deletions': commit.deletions, 'insertions': commit.insertions}
            writer.writerow(record)
```

```python
jira = JIRA(server="https://issues.apache.org/jira")

# Get all projects viewable by anonymous users.
projects = jira.projects()
# print(projects)
issue = jira.issue("AMQ-7229")
print(issue)

def get_all_issues(jira_client, project_name, fields):
    issues = []
    i = 0
    chunk_size = 100
    while True:
        print(i)
        chunk = jira_client.search_issues(f'project = {project_name}', startAt=i, maxResults=chunk_size)
        i += chunk_size
        issues += chunk.iterable
        break
        if i >= chunk.total:
            break
    return issues

issues = get_all_issues(jira, 'ActiveMQ', ['id', 'resolution', 'status', 'issuetype'])
```

```python
jiraIssuesFile = 'jiraIssues.csv'
if not(os.path.isfile(jiraIssuesFile)) :
    with open(jiraIssuesFile, 'a') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=['id', 'key', 'status', 'issuetype'])
        writer.writeheader()                        
        for issue in issues:
            print(issue.raw['fields']['issuetype'])
            print(issue.raw['fields']['status'])
            print(issue.raw['fields']['resolution'] == 'None')
            print(issue.id)
            print(issue.key)
            print("***")
            if issue.raw['fields']['issuetype']['name'] == 'Bug':
                
                record = {'id': issue.id, 'key': issue.key, 'status': issue.raw['fields']['status']['name'], 'issuetype': issue.raw['fields']['issuetype']['name']}
                writer.writerow(record)
print(len(issues))
print(issues[0].raw['fields']['description'])
```
